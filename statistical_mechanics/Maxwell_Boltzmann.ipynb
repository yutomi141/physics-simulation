```python
import numpy as np
import matplotlib.pyplot as plt
import japanize_matplotlib  
```

```python
# 定数
k_B = 1.38e-23 #ボルツマン定数
T   = 300 #温度[K]
N_A = 6.02e23 #アボガドロ定数
m = 18.0e-3 / N_A  # 水分子の質量       

# マクスウェル速さ分布関数
def maxwell(v, m):
    return 4 * np.pi * (m / (2 * np.pi * k_B * T))**1.5 * v**2 * np.exp(-m * v**2 / (2 * k_B * T))

# 特徴的速度
v_bar = np.sqrt(8 * k_B * T / (np.pi * m))
v_rms = np.sqrt(3 * k_B * T / m)
v_mp  = np.sqrt(2 * k_B * T / m)
```

```python
# プロット用データ
v = np.linspace(0, 1800, 1000)
f = maxwell(v, m)

plt.figure(figsize=(9, 6))
plt.plot(v, f, color='steelblue', lw=2, label='確率密度関数 $f(v)$')

# 垂直線とラベルの描画
plt.axvline(v_mp,  color='red',    ls='--', label=rf'最確速さ $v_{{mp}} = {v_mp:.0f}$ m/s')
plt.axvline(v_bar, color='orange', ls='--', label=rf'平均速さ $\bar{{v}} = {v_bar:.0f}$ m/s')
plt.axvline(v_rms, color='green',  ls='--', label=rf'二乗平均速度 $v_{{rms}} = {v_rms:.0f}$ m/s')

# ラベル設定
plt.xlabel('速さ $v$ m/s')
plt.ylabel('確率密度 $f(v)$ s/m')
plt.title(f'水分子の速度分布（マクスウェル-ボルツマン分布） $T = {T}$ K')

plt.legend()
plt.tight_layout()

plt.show()
```
